```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from scipy.stats import chisquare
```

```python
def f(x):
    return 4/5 * (1 + x - x**3)

def chi2(observed, expected, N):
    """Oblicza statystykę chi-kwadrat dla danych obserwowanych i oczekiwanych.
    UWAGA: observed musi być tablicą zliczeń, a expected tablicą wartości oczekiwanych dla każdego przedziału.
    N to liczba próbek, która została użyta do wygenerowania danych obserwowanych.
    """
    return np.sum((observed - expected*N)**2 / (expected*N))

expected = pd.read_csv("../results/wyniki_expected.csv", header=None).values.flatten()
```

```python
df = pd.read_csv("../results/wyniki_complex.csv", header=None)

observed = df.values.flatten()

print("Chi2:", chi2(observed, expected, 1e6))
print("P-value:", chisquare(observed, f_exp=expected*1e6)[1])

y = observed / 1e6 / 1e-1
y = np.append(y, 0)

steps = np.linspace(0, 1, 11)

x = np.linspace(0, 1, 100)

plt.figure(dpi=200)

plt.step(steps, y, where="post")
plt.plot(x, f(x), color="red")

plt.text(0.2, 0.9, f"$\chi^2 = ${chi2(observed, expected, 1e6):.2f}\n$p = ${chisquare(observed, f_exp=expected*1e6)[1]:.3f}",
         ha='center', va='center', transform=plt.gca().transAxes)

plt.title("Rozkład złożony")

plt.grid()
plt.xlim(0, 1)
plt.ylim(0.8, 1.15)

np.sum(y * 0.1)
```
